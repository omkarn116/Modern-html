3. Define accessibility in the context of web development. Discuss why it's essential to create accessible websites and how it benefits different user groups.

 **What is Accessibility in Web Development?**
-  Accessibility in web development refers to the practice of designing and building websites, applications, and digital content that can be used by everyone, including people with disabilities. This means ensuring that users with varying physical, cognitive, and sensory abilities can perceive, understand, navigate, and interact with web content effectively.
Accessibility is often guided by standards such as the Web Content Accessibility Guidelines (WCAG), which provide recommendations to make web content more accessible.

 **Why Is Web Accessibility Essential?**
- **Inclusivity and Equal Access:**

  The internet is a vital resource for education, employment, commerce, and social interaction. Making websites accessible ensures that no one is excluded from these opportunities, regardless of disability.

- **Legal Compliance:**

 Many countries have laws requiring digital accessibility (e.g., the Americans with Disabilities Act in the U.S., the European Accessibility Act). Non-compliance can lead to legal consequences, including lawsuits and fines.

- **Improved User Experience for Everyone:**

 Accessible design often enhances usability for all users—not just those with disabilities. For example, captions help not only the deaf but also people in noisy environments, and keyboard navigation helps power users as well as people with motor impairments.

- **Broader Audience Reach:**

 Making your website accessible increases your potential audience, including aging populations and users with temporary impairments (e.g., a broken arm, eye surgery).

- **SEO Benefits**

 Many accessibility best practices overlap with search engine optimization techniques, such as using semantic HTML, alt text for images, and clear headings. This can improve your site’s visibility in search engines.

- **Corporate Social Responsibility (CSR):**

 Creating accessible web content demonstrates ethical responsibility and commitment to diversity, equity, and inclusion.

 **Who Benefits from Web Accessibility?**
- People with disabilities (e.g., visual, auditory, motor, cognitive, neurological)

- Older adults who may experience age-related impairments

- People with temporary disabilities (e.g., injury, recovery from surgery)

- Users in challenging environments (e.g., low light, loud settings, poor internet connections)

- Non-native language speakers who benefit from clear, simple content

- Mobile users navigating with different devices or inputs





4. List any 3 ways which help us in improving the accessibility of HTML.
- **Use Semantic HTML Elements**

 Tags like <header>, <nav>, <main>, <section>, and <footer> help screen readers and other assistive technologies understand the structure of the page.

- **Add alt Text to Images**

 Use the alt attribute to provide descriptive text for images so that users relying on screen readers know what the image represents.

- **Use ARIA (Accessible Rich Internet Applications) Attributes Wisely**

 ARIA roles, states, and properties (like aria-label, aria-hidden, role="button") help improve accessibility for complex UI elements, but they should be used only when native HTML can't provide the needed functionality.

5. Create a web page that highlights the features of HTML5. Use appropriate semantic tags to structure the content and showcase at least three key features of HTML5 with explanations.

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>HTML5 Features</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            margin: 0;
            padding: 0;
            line-height: 1.6;
        }
        header, footer {
            background-color: #333;
            color: white;
            padding: 1em 2em;
            text-align: center;
        }
        nav {
            background: #555;
            padding: 0.5em;
            text-align: center;
        }
        nav a {
            color: white;
            margin: 0 1em;
            text-decoration: none;
        }
        section {
            padding: 2em;
        }
        article {
            margin-bottom: 2em;
        }
        h2 {
            color: #333;
        }
    </style>
</head>
<body>

    <header>
        <h1>Explore HTML5 Features</h1>
        <p>Discover the powerful enhancements introduced with HTML5</p>
    </header>

    <nav>
        <a href="#video">Video Support</a>
        <a href="#canvas">Canvas Element</a>
        <a href="#forms">Improved Forms</a>
    </nav>

    <section>
        <article id="video">
            <h2>1. Native Video Support</h2>
            <p>HTML5 introduced the <code>&lt;video&gt;</code> element, allowing developers to embed videos directly into web pages without relying on external plugins like Flash.</p>
            <video controls width="320">
                <source src="sample.mp4" type="video/mp4">
                Your browser does not support the video tag.
            </video>
        </article>

        <article id="canvas">
            <h2>2. Canvas Element</h2>
            <p>The <code>&lt;canvas&gt;</code> element enables dynamic, scriptable rendering of 2D graphics. It’s commonly used for games, visualizations, and interactive graphics using JavaScript.</p>
            <canvas id="demoCanvas" width="300" height="150" style="border:1px solid #000000;"></canvas>
            <script>
                const canvas = document.getElementById('demoCanvas');
                const ctx = canvas.getContext('2d');
                ctx.fillStyle = 'blue';
                ctx.fillRect(20, 20, 100, 100);
            </script>
        </article>

        <article id="forms">
            <h2>3. Improved Form Elements</h2>
            <p>HTML5 added new input types such as <code>email</code>, <code>date</code>, and <code>range</code>, enhancing form validation and user experience without additional JavaScript.</p>
            <form>
                <label for="email">Email:</label>
                <input type="email" id="email" name="email" required><br><br>

                <label for="dob">Date of Birth:</label>
                <input type="date" id="dob" name="dob"><br><br>

                <label for="volume">Volume:</label>
                <input type="range" id="volume" name="volume" min="0" max="100"><br><br>

                <button type="submit">Submit</button>
            </form>
        </article>
    </section>

    <footer>
        <p>&copy; 2025 HTML5 Showcase. All rights reserved.</p>
    </footer>

</body>
</html>
```

6. Create a simple web page which has a table. The table must have 2 columns HTML and HTML5. The table should include a minimum of three rows describing the differences between HTML and HTML5.
-

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>HTML vs HTML5</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            margin: 40px;
            background-color: #f9f9f9;
        }
        table {
            border-collapse: collapse;
            width: 60%;
            margin: auto;
            background-color: #fff;
        }
        th, td {
            border: 1px solid #ccc;
            padding: 12px 20px;
            text-align: left;
        }
        th {
            background-color: #f2f2f2;
        }
        caption {
            margin-bottom: 10px;
            font-size: 1.2em;
            font-weight: bold;
        }
    </style>
</head>
<body>

    <table>
        <caption>Differences Between HTML and HTML5</caption>
        <tr>
            <th>HTML</th>
            <th>HTML5</th>
        </tr>
        <tr>
            <td>Does not support native audio and video embedding.</td>
            <td>Supports native audio and video embedding with &lt;audio&gt; and &lt;video&gt; tags.</td>
        </tr>
        <tr>
            <td>Relies on external plugins like Flash for multimedia content.</td>
            <td>Eliminates the need for external plugins for most multimedia content.</td>
        </tr>
        <tr>
            <td>Limited semantic elements (mostly &lt;div&gt; and &lt;span&gt;).</td>
            <td>Introduces new semantic elements like &lt;header&gt;, &lt;footer&gt;, &lt;article&gt;, and &lt;section&gt;.</td>
        </tr>
    </table>

</body>
</html>
```
